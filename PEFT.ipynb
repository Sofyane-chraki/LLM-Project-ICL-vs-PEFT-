```python
from transformers import AutoModelForCausalLM, AutoTokenizer, BitsAndBytesConfig,HfArgumentParser,TrainingArguments,pipeline, logging
from peft import LoraConfig, PeftModel, prepare_model_for_kbit_training, get_peft_model
import os, torch, wandb
from datasets import load_dataset
from datasets import load_metric
from trl import SFTTrainer
```

```python
from kaggle_secrets import UserSecretsClient
user_secrets = UserSecretsClient()
secret_hf = user_secrets.get_secret("HUGGINGFACE_TOKEN")
secret_wandb = user_secrets.get_secret("wandb")
```

```python
# Monitering the LLM
wandb.login(key = secret_wandb)
run = wandb.init(
    project='Fine tuning mistral 7B', 
    job_type="training", 
    anonymous="allow"
)
```

```python
base_model = "/kaggle/input/mistral/pytorch/7b-v0.1-hf/1"
dataset_name = "/kaggle/input/dataset"
new_model = "mistral_7b_yann"
model_dir = "/kaggle/working/"
```

```python
#Importing the dataset
dataset = load_dataset(dataset_name, split="train")
dataset["text"][100]
```

```python
# Load base model(Mistral 7B)
bnb_config = BitsAndBytesConfig(  
    load_in_4bit= True,
    bnb_4bit_quant_type= "nf4",
    bnb_4bit_compute_dtype= torch.bfloat16,
    bnb_4bit_use_double_quant= False,
)
model = AutoModelForCausalLM.from_pretrained(
        base_model,
        quantization_config=bnb_config,
        torch_dtype=torch.bfloat16,  # Pass torch.bfloat16 here
        device_map="auto",
        trust_remote_code=True,
)
model.config.use_cache = False # silence the warnings. Please re-enable for inference!
model.config.pretraining_tp = 1
model.gradient_checkpointing_enable()

# Load tokenizer
tokenizer = AutoTokenizer.from_pretrained(base_model, trust_remote_code=True)
tokenizer.padding_side = 'right'
tokenizer.pad_token = tokenizer.eos_token
tokenizer.add_eos_token = True
tokenizer.add_bos_token, tokenizer.add_eos_token
```

```python
#Adding the adapters in the layers
model = prepare_model_for_kbit_training(model)
peft_config = LoraConfig(
    lora_alpha=16,
    lora_dropout=0.1,
    r=64,
    bias="none",
    task_type="CAUSAL_LM",
    target_modules=["q_proj", "k_proj", "v_proj", "o_proj","gate_proj"]
)
model = get_peft_model(model, peft_config)
```

```python
#Hyperparamter
training_arguments = TrainingArguments(
    output_dir="./results",
    num_train_epochs=10,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=1,
    optim="paged_adamw_32bit",
    save_steps=25,
    logging_steps=25,
    learning_rate=2e-4,
    weight_decay=0.001,
    fp16=False,
    bf16=False,
    max_grad_norm=0.3,
    max_steps=-1,
    warmup_ratio=0.03,
    group_by_length=True,
    lr_scheduler_type="constant",
    report_to="wandb"
)
```

```python
# Setting sft parameters
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=peft_config,
    max_seq_length= None,
    dataset_text_field="text",
    tokenizer=tokenizer,
    args=training_arguments,
    packing= False,
)
```

```python
if os.path.exists(os.path.join(model_dir, new_model)):
    # Load the trained model
    model = AutoModelForCausalLM.from_pretrained(os.path.join(model_dir, new_model))
else:
    # Train the model
    trainer.train()
    # After training, save the trained model
    if not os.path.exists(model_dir):
        os.makedirs(model_dir)
    model.save_pretrained(os.path.join(model_dir, new_model))
```

```python
# Save the fine-tuned model
trainer.model.save_pretrained(new_model)
wandb.finish()
model.config.use_cache = True
model.eval()
```

```python
try:
    trainer.model.push_to_hub(new_model, use_temp_dir=False)
except:
    print("An exception occurred")
```

```python
# Set logging verbosity to avoid extraneous output
logging.set_verbosity(logging.CRITICAL)

# Define the prompt
prompt = input("Enter your question: ")

# Generate text using the pipeline
pipe = pipeline(task="text-generation", model=model, tokenizer=tokenizer, max_length=200)
result = pipe(f"<s>[INST] {prompt} [/INST]")

# Extracting question and answer pairs from the generated text
generated_text = result[0]['generated_text']
question_answer_pairs = generated_text.split('[INST]')[1:]  # Exclude the first element (not a pair)

# Initialize variables to store the first question-answer pair
first_question = ""
first_answer = ""

# Iterate through question-answer pairs
for pair in question_answer_pairs:
    pair_parts = pair.split('[/INST]', 1)  # Split only once
    if len(pair_parts) == 2:
        first_question, first_answer = pair_parts
        break  # Break after finding the first valid pair

# Print the first question and its corresponding answer
print("Question:", first_question.strip())
print("Answer:", first_answer.strip())

# Calculate ROUGE-L metric for the first pair only
reference = [first_question.strip()]
hypothesis = [first_answer.strip()]

rouge_metric = load_metric("rouge")
rouge_scores = rouge_metric.compute(predictions=hypothesis, references=reference)
print("ROUGE-L Score:", rouge_scores["rougeL"].mid.fmeasure)
```
